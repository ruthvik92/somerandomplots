# Plots

```python
def plot_bode(frequencies, input_amplitudes, output_amplitudes, label, fig, ax, xticks, title, offset=0):
    magnitude = 20 * np.log10(np.array(output_amplitudes) / np.array(input_amplitudes)) + offset

    ax.set_xscale('log')
    ax.plot(frequencies, magnitude, label=label)
    ax.set_xlabel('Frequency (Hz)')
    ax.set_ylabel('Magnitude (dB)')
    ax.tick_params('y')
    ax.grid(True)

    ax.set_title(title)
    ax.legend()
    
    for xtick in xticks:
        ax.axvline(xtick, color='gray', linestyle='--', linewidth=2)

    ax.set_xticks(xticks, minor=False)
    ax.set_xticklabels(xticks, rotation=90, ha='right')

# zero_db left
frequencies = [25, 30, 40, 50, 100, 200, 1000, 5000, 10000, 12500, 15000, 17500, 20000]  # Frequencies in Hz
input_amplitudes = zero_db_input_data["Left"]  # Input amplitudes
output_amplitudes =   zero_db_output_data['Left']# Output amplitudes
label = '0db_Left'  # Label for the plot
fig, ax = plt.subplots(figsize=(12, 8))
plot_bode(frequencies, input_amplitudes, output_amplitudes, label, fig, ax, xticks=frequencies, title='Nakamichi 582 Bode Plot')


# zero_db right
input_amplitudes = zero_db_input_data["Right"]  # Input amplitudes
output_amplitudes =   zero_db_output_data['Right']# Output amplitudes
label = '0db_Right'  # Label for the plot
plot_bode(frequencies, input_amplitudes, output_amplitudes, label, fig, ax, xticks=frequencies, title='Nakamichi 582 Bode Plot')


# neg_five_db left
frequencies = [25, 30, 40, 50, 100, 200, 1000, 5000, 10000, 12500, 15000, 17500, 20000]  # Frequencies in Hz
input_amplitudes = neg_five_db_input_data["Left"]  # Input amplitudes
output_amplitudes =   neg_five_db_output_data['Left']# Output amplitudes
label = '-5db_Left'  # Label for the plot
plot_bode(frequencies, input_amplitudes, output_amplitudes, label, fig, ax, xticks=frequencies, title='Nakamichi 582 Bode Plot', offset=-5)


# neg_five_db right
frequencies = [25, 30, 40, 50, 100, 200, 1000, 5000, 10000, 12500, 15000, 17500, 20000]  # Frequencies in Hz
input_amplitudes = neg_five_db_input_data["Right"]  # Input amplitudes
output_amplitudes =   neg_five_db_output_data['Right']# Output amplitudes
label = '-5db_Left'  # Label for the plot
plot_bode(frequencies, input_amplitudes, output_amplitudes, label, fig, ax, xticks=frequencies, title='Nakamichi 582 Bode Plot', offset=-5)


plt.show()
```
